# Basic Statistics Using Python

## Assignment Objective
This Python program asks the user to enter a list of numbers, stores the numbers in a **List** and a **Dictionary**, and calculates basic statistical values.

The program performs the following operations:
- Calculates the sum
- Calculates the average
- Finds the maximum value
- Finds the minimum value
- Counts the frequency of each number


## Instructions
Enter numbers separated by spaces or commas.

**Example:**

`10 20 30 20 10`

or

`10, 20, 30, 20, 10`


## Program Setup

Run this cell first. It imports the required display tools and defines the function used to process the numbers.


````python
# Import tools used to display Markdown in Google Colab.
from IPython.display import display, Markdown


# This function processes the numbers and returns the required results.
def process_numbers(number_list):
    # Create an empty dictionary to store number frequencies.
    frequency_dict = {}

    # Count how many times each number appears.
    for number in number_list:
        if number in frequency_dict:
            frequency_dict[number] += 1
        else:
            frequency_dict[number] = 1

    # Calculate the required statistics.
    total = sum(number_list)
    average = total / len(number_list)
    maximum = max(number_list)
    minimum = min(number_list)

    return frequency_dict, total, average, maximum, minimum
````

## Main Program

Run this cell after the setup cell, then enter numbers separated by spaces or commas.


````python
# Ask the user to enter numbers separated by spaces or commas.
user_input = input("Enter numbers separated by spaces or commas: ")

# Replace commas with spaces, then split the input into separate values.
number_strings = user_input.replace(",", " ").split()

try:
    # Convert each entered value into a floating-point number.
    number_list = [float(value) for value in number_strings]

    # Check that the user entered at least one number.
    if not number_list:
        display(Markdown("**Error:** Please enter at least one number."))

    else:
        # Process the numbers by calling the function.
        frequency_dict, total, average, maximum, minimum = process_numbers(
            number_list
        )

        # Prepare the main results as a Markdown table.
        results_output = f"""
## Statistical Results

| Statistic | Result |
|:--|:--|
| List of numbers | `{number_list}` |
| Frequency dictionary | `{frequency_dict}` |
| Sum | {total} |
| Average | {average:.2f} |
| Maximum value | {maximum} |
| Minimum value | {minimum} |
"""

        # Display the results.
        display(Markdown(results_output))

except ValueError:
    # This message appears when a non-numeric value is entered.
    display(Markdown("**Error:** Please enter numbers only."))
````

## Bonus Results

This optional section sorts the numbers and identifies the most frequent number and its number of occurrences.

Run it only after running the main program cell.


````python
# Optional Bonus Features
try:
    # Check that the main program created a non-empty number list.
    if not number_list:
        display(Markdown("**Error:** Please enter at least one number in the main program."))

    else:
        # Sort the numbers in ascending order.
        sorted_numbers = sorted(number_list)

        # Find the number with the highest frequency.
        most_frequent_number = max(
            frequency_dict,
            key=frequency_dict.get
        )
        highest_frequency = frequency_dict[most_frequent_number]

        # Prepare the bonus results as a Markdown table.
        bonus_output = f"""
## Bonus Results

| Bonus Feature | Result |
|:--|:--|
| Numbers in ascending order | `{sorted_numbers}` |
| Most frequent number | {most_frequent_number} |
| Number of occurrences | {highest_frequency} |
"""

        # Display the bonus results.
        display(Markdown(bonus_output))

except NameError:
    display(Markdown("**Error:** Please run the setup and main program cells first."))
````

## Extra Feature

As an additional feature, the program allows the user to choose which statistical operation to display.

The available operations are:

* Calculate the sum
* Calculate the average
* Find the maximum value
* Find the minimum value
* Display the frequency dictionary
* Display all statistical results

Run this cell after the main program.


````python
# Extra Feature: Operation Selection
# This section allows the user to choose a statistical operation
# using the numbers entered in the main program.

try:
    # Make sure the number list contains values.
    if not number_list:
        print("Error: The number list is empty.")

    else:
        # Display the available statistical operations.
        print("\n--- Choose a Statistical Operation ---")
        print("1. Calculate the sum")
        print("2. Calculate the average")
        print("3. Find the maximum value")
        print("4. Find the minimum value")
        print("5. Display the frequency dictionary")
        print("6. Display all statistical results")

        # Ask the user to select an operation.
        choice = input("Enter your choice from 1 to 6: ")

        # Perform the selected operation.
        if choice == "1":
            print("Sum:", total)

        elif choice == "2":
            print("Average:", round(average, 2))

        elif choice == "3":
            print("Maximum value:", maximum)

        elif choice == "4":
            print("Minimum value:", minimum)

        elif choice == "5":
            print("Frequency dictionary:", frequency_dict)

        elif choice == "6":
            print("\n--- All Statistical Results ---")
            print("List of numbers:", number_list)
            print("Frequency dictionary:", frequency_dict)
            print("Sum:", total)
            print("Average:", round(average, 2))
            print("Maximum value:", maximum)
            print("Minimum value:", minimum)

        else:
            print("Error: Please enter a number from 1 to 6.")

except NameError:
    print("Error: Please run the setup and main program cells first.")
````
