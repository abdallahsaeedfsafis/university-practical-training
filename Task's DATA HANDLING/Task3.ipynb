# Titanic Dataset Cleaning and Enhancement

This notebook cleans and enhances the Titanic dataset using Python and Pandas in Google Colab.  
The goal is to prepare the dataset for real-world data science, visualization, and AI modeling tasks.

```python
# Import the Pandas library for data handling and analysis
import pandas as pd

# Store the Titanic dataset URL in a variable
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

# Read the CSV file from the URL and store it in a DataFrame
titanic_df = pd.read_csv(url)

# Display the first 5 rows of the dataset to check that the data loaded correctly
display(titanic_df.head())
```

```python
# Display the number of rows and columns in the dataset
print("Dataset shape:")
print(titanic_df.shape)

# Display the column names in the dataset
print("\nColumn names:")
print(titanic_df.columns)

# Display general information about the dataset
# This includes column names, data types, and non-missing values
print("\nDataset information:")
titanic_df.info()

# Display the number of missing values in each column
print("\nMissing values in each column:")
display(titanic_df.isnull().sum())

# Display the number of duplicated rows in the dataset
print("\nNumber of duplicated rows:")
print(titanic_df.duplicated().sum())

# Display basic summary statistics for numeric columns
print("\nSummary statistics:")
display(titanic_df.describe())
```

```python
# Store the number of rows before cleaning
rows_before_cleaning = titanic_df.shape[0]

# Store the number of duplicated rows before removing duplicates
duplicates_before = titanic_df.duplicated().sum()

# Store missing values before cleaning
missing_before = titanic_df.isnull().sum()

# Calculate the median value of the Age column
# The median is used because it is less affected by extreme values than the mean
age_median = titanic_df["Age"].median()

# Fill missing Age values with the median age
titanic_df["Age"] = titanic_df["Age"].fillna(age_median)

# Calculate the most frequent value in the Embarked column
# The mode is suitable for categorical columns
embarked_mode = titanic_df["Embarked"].mode()[0]

# Fill missing Embarked values with the most frequent value
titanic_df["Embarked"] = titanic_df["Embarked"].fillna(embarked_mode)

# Drop the Cabin column because it has many missing values
titanic_df = titanic_df.drop(columns=["Cabin"])

# Remove duplicated rows to ensure data integrity
titanic_df = titanic_df.drop_duplicates()

# Standardize passenger names using title case
titanic_df["Name"] = titanic_df["Name"].str.title()

# Standardize the Sex column by removing spaces and converting values to lowercase
titanic_df["Sex"] = titanic_df["Sex"].str.strip().str.lower()

# Standardize the Embarked column by removing spaces and converting values to uppercase
titanic_df["Embarked"] = titanic_df["Embarked"].str.strip().str.upper()

# Store the number of rows after cleaning
rows_after_cleaning = titanic_df.shape[0]

# Store missing values after cleaning
missing_after = titanic_df.isnull().sum()

# Display cleaning results
print("Rows before cleaning:", rows_before_cleaning)
print("Rows after cleaning:", rows_after_cleaning)
print("Duplicated rows before cleaning:", duplicates_before)
print("Median Age used for imputation:", age_median)
print("Mode Embarked used for imputation:", embarked_mode)

print("\nMissing values before cleaning:")
display(missing_before)

print("\nMissing values after cleaning:")
display(missing_after)

# Display the cleaned dataset
display(titanic_df.head())
```

```python
# Create a new column called FamilySize
# It represents the total number of people traveling together
# SibSp = siblings/spouses, Parch = parents/children, +1 = the passenger himself/herself
titanic_df["FamilySize"] = titanic_df["SibSp"] + titanic_df["Parch"] + 1

# Define a function to classify passengers based on their family/travel group size
def classify_travel_group(size):
    # If the passenger is alone, classify as Solo
    if size == 1:
        return "Solo"

    # If the group size is 2 or 3, classify as Small
    elif size <= 3:
        return "Small"

    # If the group size is 4 or more, classify as Large
    else:
        return "Large"

# Apply the function to create a new TravelGroup column
titanic_df["TravelGroup"] = titanic_df["FamilySize"].apply(classify_travel_group)

# Create AgeGroup to make the Age column easier to analyze
# This groups passengers into age categories
titanic_df["AgeGroup"] = pd.cut(
    titanic_df["Age"],
    bins=[0, 12, 19, 59, 100],
    labels=["Child", "Teen", "Adult", "Senior"]
)

# Display selected columns to check the new engineered features
display(titanic_df[["Age", "AgeGroup", "SibSp", "Parch", "FamilySize", "TravelGroup"]].head(10))
```

```python
# Store the original Fare values before handling outliers
# This helps us compare the Fare column before and after outlier treatment
titanic_df["OriginalFare"] = titanic_df["Fare"]

# Calculate the first quartile Q1 for the Fare column
# Q1 represents the value below which 25% of the Fare values fall
Q1 = titanic_df["Fare"].quantile(0.25)

# Calculate the third quartile Q3 for the Fare column
# Q3 represents the value below which 75% of the Fare values fall
Q3 = titanic_df["Fare"].quantile(0.75)

# Calculate the Interquartile Range
# IQR measures the spread of the middle 50% of the Fare values
IQR = Q3 - Q1

# Calculate the lower bound for acceptable Fare values
lower_bound = Q1 - 1.5 * IQR

# Calculate the upper bound for acceptable Fare values
upper_bound = Q3 + 1.5 * IQR

# Count how many Fare values are considered outliers before handling them
fare_outliers_before = titanic_df[
    (titanic_df["Fare"] < lower_bound) | (titanic_df["Fare"] > upper_bound)
].shape[0]

# Cap Fare values using the lower and upper bounds
# Values below the lower bound become equal to the lower bound
# Values above the upper bound become equal to the upper bound
titanic_df["Fare_Capped"] = titanic_df["Fare"].clip(
    lower=lower_bound,
    upper=upper_bound
)

# Count how many Fare values are still outside the bounds after capping
fare_outliers_after = titanic_df[
    (titanic_df["Fare_Capped"] < lower_bound) | (titanic_df["Fare_Capped"] > upper_bound)
].shape[0]

# Display the IQR values and outlier handling results
print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("Lower bound:", lower_bound)
print("Upper bound:", upper_bound)
print("Fare outliers before handling:", fare_outliers_before)
print("Fare outliers after handling:", fare_outliers_after)

# Display selected columns to compare original Fare and capped Fare
display(titanic_df[["OriginalFare", "Fare_Capped"]].head(10))
```

## Cleaning and Enhancement Documentation

In this step, the Titanic dataset was cleaned and enhanced to make it more reliable for data analysis and AI modeling.

The cleaning process included handling missing values, removing duplicates, standardizing text formats, creating new useful features, and handling Fare outliers.

Missing Age values were filled using the median age because the median is less affected by extreme values.

Missing Embarked values were filled using the mode because Embarked is a categorical column.

The Cabin column was dropped because it had too many missing values.

Duplicate rows were checked and removed to ensure data integrity.

Passenger names were standardized using title case, Sex values were converted to lowercase, and Embarked values were converted to uppercase.

New features were created:
- FamilySize: calculated using SibSp + Parch + 1
- TravelGroup: classifies passengers as Solo, Small, or Large
- AgeGroup: classifies passengers as Child, Teen, Adult, or Senior

Fare outliers were handled using the IQR method. The lower bound was calculated as Q1 - 1.5 * IQR, and the upper bound was calculated as Q3 + 1.5 * IQR. Values outside these bounds were capped to the nearest acceptable boundary.

```python
# Create a final summary dictionary to document the cleaning results
cleaning_summary = {
    # Number of rows before cleaning
    "Rows before cleaning": rows_before_cleaning,

    # Number of rows after cleaning
    "Rows after cleaning": rows_after_cleaning,

    # Number of duplicated rows found before cleaning
    "Duplicated rows before cleaning": duplicates_before,

    # Median value used to fill missing Age values
    "Median Age used": age_median,

    # Mode value used to fill missing Embarked values
    "Mode Embarked used": embarked_mode,

    # Number of Fare outliers before applying IQR capping
    "Fare outliers before handling": fare_outliers_before,

    # Number of Fare outliers after applying IQR capping
    "Fare outliers after handling": fare_outliers_after,

    # Mean Fare after capping outliers
    "Mean capped Fare": titanic_df["Fare_Capped"].mean(),

    # Maximum family/travel group size
    "Maximum Family Size": titanic_df["FamilySize"].max()
}

# Display the final cleaned and enhanced dataset
print("Final cleaned and enhanced Titanic dataset:")
display(titanic_df.head(10))

# Display the final cleaning summary
print("Final cleaning summary:")
display(cleaning_summary)

# Display the number of passengers in each TravelGroup category
print("Passenger count by TravelGroup:")
display(titanic_df["TravelGroup"].value_counts())

# Display the number of passengers in each AgeGroup category
print("Passenger count by AgeGroup:")
display(titanic_df["AgeGroup"].value_counts())
```

```python
# Create a cleaner final summary dictionary using regular Python data types
# This makes the output easier to read and present

final_summary = {
    "Rows before cleaning": int(rows_before_cleaning),
    "Rows after cleaning": int(rows_after_cleaning),
    "Duplicated rows before cleaning": int(duplicates_before),
    "Median Age used": float(age_median),
    "Mode Embarked used": str(embarked_mode),
    "Fare outliers before handling": int(fare_outliers_before),
    "Fare outliers after handling": int(fare_outliers_after),
    "Mean capped Fare": round(float(titanic_df["Fare_Capped"].mean()), 2),
    "Maximum Family Size": int(titanic_df["FamilySize"].max())
}

# Display the clean final summary
display(final_summary)
```
